```python
def findThreeLargestNumbers(array): # O(n) time | O(1) space 
  threeLargest=[None, None, None]

  for num in array:
    updateLargest(threeLargest, num)
  
  return threeLargest 

def updateLargest(threeLargest, num):

  if threeLargest[2] is None or num > threeLargest[2]:
    shiftAndUpdate(threeLargest, num, 2)
  
  elif threeLargest[1] is None or num> threeLargest[1]:
    shiftAndUpdate(threeLargest, num, 1) 
  
  elif threeLargest[0] is None or num > threeLargest[0]:
    shiftAndUpdate(threeLargest, num, 0) 

def shiftAndUpdate(array, num, idx):

  for i in range(idx+1):
    if i == idx:
      array[i] = num
    
    else:
      array[i]=array[i+1]


```

```python
def bubbleSort(array):   # O(n pow 2) time | O(1) space 
  isSorted=False 
  counter=0
  while not isSorted:
    isSorted=True 
    for i in range(len(array) -1- counter):# by the end of first iteration one element at it perfect postion at end
      if array[i] > array[i+1]:
        swap(i, i+1, array)
        isSorted=False 
    counter+=1
  return array 


def swap(i, j, array):
  array[i], array[j]= array[j], array[i]

```

```python
# LRU 
# most recently used Linked nodes will be nead the tail 
# least recent nodes will be near the head 
# we will remove the head node 
# for checking contains, will shift the key, value to the tail (MOST RECENTLY USED)
# for eviction we will use least-recently-used( head )
class LinkedNode(object): 
  def __init__(self, val, key):
    self.val= val 
    self.key= key
    self.next= None 
    self.prev= None 


class LruCache: 

  def __init__(self, capacity):
    self.capacity= capacity 
    
    self.lookup= {}                        # hashmap
    self.dummy= LinkedNode(0,0)            # doubly linked list
    
    self.head= self.dummy.next 
    self.tail= self.dummy.next 

  
  def remove_head_node(self):
    if not self.head: 
      return 
    prev= self.head 
    self.head= self.head.next 
    if self.head: 
      self.head.prev= None 
    
    del prev 
  
  def append_new_node(self, new_node):
    """ add a new node to the tail end""" 

    if not self.tail: 
      self.head= self.tail= new_node  
    
    else: 
      self.tail.next= new_node     # referencing new node
      new_node.prev= self.tail     # referencing new node's prev to tail node 
      self.tail= self.tail.next    # updating tail node
    
   def unlink_cur_node(self, node): 
     """ unLink current Linked node""" 
     if self.head is node: 
       self.head = node.next 
       if node.next: 
         node.next.prev= None 
       return 
     
     # removing the node from somewher in the middle, updat pointers 
     prev, nex= node.prev, node.next 
     prev.next= nex 
     nex.prev= prev 

  def get(self, key):
    if key not in self.lookup:   # what if called from funtion outside this class 
      return -1 
    
    node=self.lookup[key]     # get corresponding node reference

    if node is not self.tail:
      self.unlink_cur_node(node)  # unlink the node from current Postion, and move it to the tail(as this is most recently used one)
      self.append_new_node(node) 
    
    return node.val 
   
  def put(self, key, value):

    if key in self.lookup:           # if key in hashmap
      self.lookup[key].val= value    # updating the value at this key
      self.get(key)     # now this has been referenced, shift this to end(check for same) if not append to tail
      return 
  

    if len(self.lookup) == self.capacity:   # if list is full 
      # remove head node and correspond key 

      self.lookup.pop(self.head.key)       # remove the least recently used one (head) to make room 
      self.remove_head_node()            
      

    new_node= LinkedNode(val=value, key=key)   # create a new DLL node 
    self.lookup[key]= new_node                 # updating dict for this node 
    self.append_new_node(new_node)             # append this node to the tail 
```

```python
def iterativeInOrderTraversal(tree, callback):    # O(n) time  | O(1) space 

  previousNode= None 
  currentNode = tree 
  while currentNode is not None: 
    if previousNode is None or previousNode== currentNode.parent: 
      if currentNode.left is not None: 
        nextNode= currentNode.left 
      
      else: 
        callback(currentNode) 
        nextNode= currentNode.right if currentNode.right is not None else currentNode.parent 

    elif previousNode== currentNode.left: 
      callback(currentNode) 
      nextNode= currentNode.right if currentNode.right is not None else currentNode.parent 

    else: 
      nextNode = currentNode.parent 

    previousNode= currentNode 
    currentNode= nextNode 
```

```python
def knuthMorrisPrattAlgorithm(string, substring):  # O(n+m) | O(m) space  
  pattern= buildPattern(substring) 
  return doesMatch(string, substring, pattern) 



def buildPattern(substring): 
  pattern= [-1 for i in substring] 
  j=0 
  i=1 
  while i< len(substring): 

    if substring[i] == substring[j]: 
      pattern[i] = j 
      i+=1 
      j+=1 

    elif j>0:  # no pattern has been identified in substring 
      j= pattern[j-1]+1  # reset j value to a previous index from the lookup table 

    else: 
      i+=1 
  return pattern 


def doesMatch(string, substring, pattern): 
  i=0 
  j=0 
  while i+ len(substring) -j <= len(string):  # sub string 98 characters, string 99 characters
    if string[i] == substring[j]: 
      if j== len(substring)-1:  # if j is last character in substring, and we have reached the end, we return 
        return True 

      i+=1 
      j+=1 
    
    elif j>0: 
      j= pattern[j-1] +1 
    

    else: 
      i+=1
  
  return False 


```
